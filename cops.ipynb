### Nombre de policiers pour 100 000 habitants par pays (en python)

#### Population en 2007

```python
import numpy as np
```

```python
europpop = pd.read_csv('pop1.csv')
eurofilt=europpop
eurofilt = eurofilt[eurofilt['Value']!=':']

eurofilt['Value'] = np.array(int(str(i).replace('.','')) for i in eurofilt['Value'])
eurofilt['Value']= eurofilt['Value'].astype("int")
#eurofilt['Value'] 
eurofilt
```

```python
 countriespop = ["Allemagne incluant l'ancienne RDA", 'Autriche', 'Belgique', 'Espagne', 'Finlande', 'France',
              'Irlande', 'Italie', 'Luxembourg', 'Pays-Bas', 'Portugal', 'Gr�ce',
              'Slov�nie', 'Chypre', 'Malte', 'Slovaquie', 'Estonie', 'Lettonie', 'Lituanie']
```

```python
europexpect = {}

for i in range(len(countriespop)):
    df = eurofilt[['GEO','TIME','Value']][(eurofilt['GEO']==countriespop[i])]

    X = np.array(df['Value']) 
    y = np.array(df['TIME'])

    P = np.polyfit(y,X,1)
    
    infivety= 2050 
    infourty =2040
    inseven=2007
    resfiv = int(P[0]*infivety+P[1])
    resfour =int(P[0]*infourty+P[1])
    ressev =int(P[0]*inseven+P[1])
    
    target={'name':countriespop[i],'2050':resfiv,'2040':resfour,'2007':ressev}
    
    #europexpect[countries[i]]=[resfiv,resfour]
    europexpect[i]=target
europexpect
```

##### Policiers pour 100 000 habitant en 2007

```python
copzerosev=cop[cop['TIME']==2007]
copzerosev['Value']=copzerosev['Value'].astype("int")
copzerosev= copzerosev[copzerosev['GEO'].isin(countries)]
copzerosev
```

```python
copsres=[]
for i in range(19):
    temp=0
    temp = copzerosev['Value'][copzerosev['GEO']==countries[i]].values[0]
    copsres.append(temp)
copsres
```

```python
len(copsres)
```

```python
for i in range(19):
    res=0
    #print(europexpect[i]['2007'])
    #print(copsres[i]*100000)
    res = round (copsres[i]/europexpect[i]['2007']*100000,2)
    europexpect[i]['undredk']=res
europexpect
```

```python
import json

class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NpEncoder, self).default(obj)

with open('eu19_cops.json', 'w') as f:
    json.dump(list_resp, f, cls=NpEncoder)
```
