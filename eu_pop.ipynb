```python
# Data for plotting
plt.figure()

#BeginFor
for i in range(len(countries)):
    df = eurofilt[['GEO','TIME','Value']][(eurofilt['GEO']==countries[i])]

    X = np.array(df['Value']) 
    y = np.array(df['TIME'])

    P = np.polyfit(y,X,1)
    t = df['Value']
    s = df['TIME']

    plt.plot(s,t,label=countries[i])
    plt.plot(s,P[0]*s+P[1],label=countries[i] +' tendance')
#EndFor

plt.title('Pop evolution in eu zone : 19')
plt.xlabel('year')
plt.ylabel('pop num')
plt.legend()

plt.show()
```

### Expect in 2050 and 2040 (cf :intwenty= 2020 res = int(P[0]*intwenty+P[1]) res)

```python
europexpect = {}

for i in range(len(countries)):
    df = eurofilt[['GEO','TIME','Value']][(eurofilt['GEO']==countries[i])]

    X = np.array(df['Value']) 
    y = np.array(df['TIME'])

    P = np.polyfit(y,X,1)
    
    infivety= 2050 
    infourty =2040
    resfiv = int(P[0]*infivety+P[1])
    resfour =int(P[0]*infourty+P[1])
    
    target={'name':countries[i],'2050':resfiv,'2040':resfour}
    
    #europexpect[countries[i]]=[resfiv,resfour]
    europexpect[i]=target
europexpect
```

## Save data in json

```python
#import json
#with open('eu19_population.json', 'w', encoding='utf-8') as f:
    #json.dumps(europexpect, indent=4)

import json

class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NpEncoder, self).default(obj)

with open('eu19_population.json', 'w') as f:
    json.dump(europexpect, f, cls=NpEncoder)
```
